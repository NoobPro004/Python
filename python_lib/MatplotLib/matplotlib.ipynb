```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.array([1,234])
y=np.array([1,20])
plt.plot(x,y)
plt.show()
```

**Defaults X Points**

```python
y=np.array([1,7,36])
plt.plot(y)
plt.show()
```

**Markers**

```python
y=np.array([1,7,36])
plt.plot(y,marker='o')
plt.show()
plt.plot(y,'o')
plt.show()
plt.plot(y,marker='o',linestyle=':',color='r')
plt.show()
plt.plot(y,'o:r')
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([23,446,67,774,47,55,45])
plt.plot(x,y)
plt.title("test graph")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
x=np.array([1,2,3,4,5,6])
y=np.array([35,42,4,3,233,65])
plt.plot(x,y,'o--r')
plt.title('Sales')

plt.subplot(1,2,2)
x=np.array([23,54,5,56])
y=np.array([9,8,7,6])
plt.plot(x,y,'d:g')
plt.title("Performace")
plt.suptitle("Overall Index")
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([234,4,7,6,87,899,67])
size=np.array([10,20,34,54,6,57,66])
plt.scatter(x,y,s=size,alpha=0.5)
plt.show()
```

**Bar Graph**

```python
x=np.array(['A','B','C','D','E'])
y=np.array([1,2,3,4,5])
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,color='hotpink',width=0.5)
plt.show()
```

**Histogram**

```python
x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
x=np.array([23,45,54,7,96])
labels=['apple','oranges','grapes','mushroom','banana']
exp=[0.1,0.2,0.3,0.2,0.2]
plt.pie(x,labels=labels,startangle=90,explode=exp,shadow=True)
plt.legend(title='Fruits')
plt.show()
```
