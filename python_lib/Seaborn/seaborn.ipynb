```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'])
plt.plot()
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'])
sns.set()
plt.plot()
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'])
sns.set_style('darkgrid',{'grid.color':'0.5'})
sns.despine()
plt.plot()
```

```python
sns.palplot(sns.color_palette('dark',10)) #170
```

```python
sns.relplot(data=df,x='sepal_length',y='sepal_width',hue='species')
```

```python
sns.relplot(data=df,x='sepal_length',y='sepal_width',hue='species',kind='line')
sns.relplot(data=df,x='sepal_length',y='sepal_width',kind='line')
```

```python
sns.distplot(df['sepal_length'],kde=True)
```

```python
sns.barplot(data=df,x='species',y='sepal_length',ci=None)
plt.plot()
```

```python
sns.barplot(data=df,y='species',x='sepal_length',orient='h',ci=None)
plt.plot()
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
sns.countplot(data=df,x='class',hue='who',palette='magma')
plt.title("Survivors")
```

```python
df=sns.load_dataset('iris')
sns.kdeplot(data=df,x=df[df['species']=='setosa']['sepal_length'],y=df['sepal_width'],color='r',shade=True,cmap='Reds')
sns.kdeplot(data=df,x=df[df['species']=='versicolor']['sepal_length'],y=df['sepal_width'],color='b',shade=True,cmap='Blues')
```

```python
df=sns.load_dataset('titanic')
sns.boxplot(data=df[df['fare']<200],x='class',y='fare')
```

```python
df=sns.load_dataset('flights')
df.head()
df=df.pivot('year','month','passengers')
sns.heatmap(df)
```

```python
df=sns.load_dataset('iris')
sns.pairplot(df,hue='species',kind='scatter')
```
