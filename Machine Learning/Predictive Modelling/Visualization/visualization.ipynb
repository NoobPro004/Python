```python
import pandas as pd
import seaborn as sns
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/onkar-99/Exploratory-Data-Analysis-on-Retail-Store/main/SampleSuperstore.csv")
```

```python
df.head()
```

```python
plt.scatter(df['Category'],df['Sales'])
plt.xlabel("Category")
plt.ylabel("Sales")
```

```python
df.boxplot(by='Discount',column='Profit',showfliers=False,figsize=(10,6),patch_artist=True)
```

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue="Region",size=3)
plt.show()
```

```python
sns.violinplot(x='Discount',y='Profit',data=df,size=6)
```

```python
sns.FacetGrid(df,hue="Region").map(sns.distplot,'Discount').add_legend()
```

```python
plt.bar(df['Discount'],df['Sales'])
```

```python
plt.plot(df['Sales'],df['Profit'])
```
