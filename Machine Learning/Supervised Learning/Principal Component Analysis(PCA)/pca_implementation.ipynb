```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df =pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled = scalar.fit_transform(df)
```

```python
answer=cancer['target']
```

```python
answer
```

```python
features = scaled.T
cov_matrix=np.cov(features)
```

```python
values, vectors = np.linalg.eig(cov_matrix)
```

```python
vectors
```

```python
percentage=[]
for value in values:
    percentage.append((value/np.sum(values))*100)

print(percentage)
```

```python
data1 = np.dot(scaled,vectors.T[0])
data2 = np.dot(scaled,vectors.T[1])
data3 = np.dot(scaled,vectors.T[2])
```

```python
result=pd.DataFrame(data1,columns=['PC1'])
```

```python
result['PC2'] = data2
result['PC3']=data3
result['ans']=answer
```

```python
result=pd.DataFrame(result,columns=['PC1','PC2','PC3','ans'])
result.head()
```

```python
result.info()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sb.scatterplot(x=result['PC1'],y=result['PC2'],hue=result['ans'],s=200)
```

```python
new =result.drop(['ans'],axis=1)
```

```python
new.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new,answer,test_size=0.33,random_state=42)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
