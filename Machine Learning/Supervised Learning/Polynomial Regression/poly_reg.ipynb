```python
import numpy as np
import pandas as pd
```

```python
x=np.random.normal(0,1,70)*10
y=10*(x**2) + np.random.normal(-100,100,70)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lir = LinearRegression()
```

```python
lir.fit(x.reshape(-1,1),y.reshape(-1,1))
```

```python
pred=lir.predict(x.reshape(-1,1))
```

```python
plt.scatter(x,y)
plt.plot(x,pred,c='m')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x.reshape(-1,1))
```

```python
poly_lin=LinearRegression()
poly_lin.fit(x_poly,y.reshape(-1,1))
```

```python
y_pred=poly_lin.predict(x_poly)
```

```python
plt.scatter(x,y)
sorted_zip=sorted(zip(x,y_pred))
x_p,y_pred=zip(*sorted_zip)
plt.plot(x_p,y_pred,color='r')
plt.show()
```
